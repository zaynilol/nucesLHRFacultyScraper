```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "http://lhr.nu.edu.pk/faculty/"
page = requests.get(url)

print(page)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# FSC scraper
```

```python
tempFSC = soup.find('div', class_='container', id = 'fsc')
fscDivsList = tempFSC.find_all('div', class_='col-lg-3 col-md-4 col-sm-6 col-12')
```

```python
fscFrame = pd.DataFrame(columns = ['ID', 'Name', 'Designation', 'HEC Approved Supervisor', 'Email', 'Department', 'Img URL'])
# print(fscFrame)

fscFrame
```

```python
for divs in fscDivsList:
    temp = divs.find('a')
    tempID = temp.get('href')

    tempLastFour = tempID[-4:]

    designation = divs.find('p', class_='small text-center font-italic').get_text()
    splitDesig = designation.split('\n', 1)[0]

    hecApproval = "HECApprovedPhDSupervisor"
    tempHecApproved = str(divs.find('p', class_="small text-center font-italic"))
    hecApproved = ''

    for c in tempHecApproved:
        if c != ' ' and c != '\n':
            hecApproved += c

    supervisorFlag = False
    if hecApproval in hecApproved:
        supervisorFlag = True

    tempNames = divs.find('h5').get_text()
    tempEmail = divs.find('p', class_='mb-0 text-center').get_text()
    tempImageURL = divs.find('img').get('src')

    tempDictionaryFSC = {
        'ID': int(tempLastFour),
        'Name': tempNames,
        'Designation': splitDesig,
        'HEC Approved Supervisor': supervisorFlag,
        'Email': tempEmail,
        'Department': "FSC",
        'Img URL': tempImageURL
    }

    tempFSCFrame = pd.DataFrame([tempDictionaryFSC])
    fscFrame = pd.concat([fscFrame, tempFSCFrame], ignore_index=True)
```
